```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import ExtraTreesClassifier
```

```python
X_train, X_dev, y_train, y_dev = train_test_split(X, Y, test_size=0.2, random_state=2474)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_dev_scaled = scaler.transform(X_dev)
```

```python
param_grid = {
    'n_estimators': [10, 50, 150],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 4],
    'max_features': [2, 5, 10, 20, 54]
}
```

```python
grid_search_ET = GridSearchCV(
    estimator=ExtraTreesClassifier(random_state=0),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
grid_search_ET.fit(X_train_scaled, y_train)
best_params_ET = grid_search_ET.best_params_
optimal_criterion_ET = best_params_ET['criterion']
optimal_n_estimators_ET = best_params_ET['n_estimators']
optimal_min_samples_split_ET = best_params_ET['min_samples_split']
optimal_max_features_ET = best_params_ET['max_features']
```

```python
print(f'The optimal criterion is {optimal_criterion_ET}')
print(f'The optimal number of trees in the forest is {optimal_n_estimators_ET}')
print(f'The optimal minimum number of samples required to split an internal node is {optimal_min_samples_split_ET}')
print(f'The optimal maximum number of features is {optimal_max_features_ET}')
```

```python
best_ET = ExtraTreesClassifier(**best_params_ET, random_state=0)
best_ET.fit(X_train_scaled, y_train)
y_dev_pred = best_ET.predict(X_dev_scaled)
accuracy = accuracy_score(y_dev, y_dev_pred)
print(f'Accuracy: {accuracy}')
```
